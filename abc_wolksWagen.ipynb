```python
def dias_vividos_v2(lista_db,lista_mb,lista_yb,lista_dd,lista_md,lista_yd):
    aux=0
    aux2=0
    lista_dias_vividos.append("___________")
    for i in range(0,len(lista_db)):
        #print("Aux 0 -----> "+str(aux))
        aux=(lista_yd[i]-lista_yb[i])
        #print("Aux ano morte - ano nasc -----> "+str(aux))
        aux=(aux*12)+lista_md[i]-lista_mb[i]
        #print("Aux ano morte - ano nasc * 12 + mes morte -mes nasc -----> "+str(aux))
        aux=aux*31
        #print("Aux 4 -----> "+str(aux))
        aux=aux+lista_dd[i]
        #print("Aux 5 -----> "+str(aux))
        aux=aux-lista_db[i]
        #print("Aux 6 -----> "+str(aux))
        aux2=aux2+aux
    #print(str(int(len(lista_db))))
    aux2=aux2/int(len(lista_db))
    lista_total.append(aux2)
    
```
